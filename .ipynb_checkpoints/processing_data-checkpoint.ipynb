## Load and test Face Detector Model

```python
import numpy as np
import cv2
import os
```

```python
## develop a function for this
def get_face(img, net):      
    ## get heigh and width
    h, w = img.shape[0], img.shape[1]

    ## construct a blob of the image for the net
    blob = cv2.dnn.blobFromImage(img,
                                 1.0,
                                 (300, 300),
                                 (104.0, 177.0, 123.0))

    ## pass the blob through the net
    net.setInput(blob)
    detections = net.forward()
    
    if detections.shape[2] > 0: 
        ## loop in the detections
        for i in range(0, detections.shape[2]):
            ## extract the confidence
            confidence = detections[0, 0, i, 2]

            ## greater the minimun confidence
            if confidence > 0.8:
                ## get the coordinates for bounding box
                box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
                start_x, start_y, end_x, end_y = box.astype("int")

                ## ensure the boundig box
                start_x, start_y = max(0, start_x), max(0, start_y)
                end_x, end_y = min(w-1, end_x), min(h-1, end_y)

                ## extract the face ROI
                face = img[start_y:end_y, start_x:end_x]
                #face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
                face = cv2.resize(face, (224, 224))
                
                return "face", face
            
            else:
                return "no face", None
    
    else:
        return "no face", None
```

```python
## get the architecture and weights of the model
proto = "face_detector/deploy.prototxt"
weights = "face_detector/res10_300x300_ssd_iter_140000.caffemodel"

## load model using opencv
net = cv2.dnn.readNet(proto, weights)

## read random image
random_image = random.choices(list_images_with_mask, k=1)
print(random_image[0])
img = cv2.imread("dataset/with_mask/" + random_image[0])
plt.axis("off")
plt.imshow(img)
plt.show()

result, face = get_face(img, net)

if result == "face":
    plt.imshow(face)
    plt.axis("off")
    plt.show()
```

## Split Face Dataset

```python
## function to split the dataset
import os
import glob
import shutil

from random import shuffle
#!pip install PrettyTable
from prettytable import PrettyTable

def split_dataset(path):
    
    ## create folder for train, val & test
    os.mkdir("dataset_split")    
    folders = ["dataset_split/train", "dataset_split/val", "dataset_split/test"]
        
    for folder in folders:
        os.mkdir(folder) 
    
    ## list the labels of the dataset
    labels = os.listdir(path)
    #print("labels: ", labels)

    ## folders for split the dataset
    train = []
    val = []
    test = []
    
    ## create a table for show information
    t = PrettyTable(['Labels', 'Train', "Val", "Test", "Total"])

    ## navigate through each label
    for label in labels:
        #print("#"*20)
        #print("#"*5, label, "#"*5)
        #print("#"*20)

        path_label = os.path.join(path, label)

        ## list images
        images = []
        images = glob.glob(path_label + "/*.jpg")
        images.extend(glob.glob(path_label + "/*.png"))
        images.extend(glob.glob(path_label + "/*.jpeg"))

        ## shuffle images
        shuffle(images)
        #print("List of images:", images[0:2])

        ## count the numbers of images
        number_images = len(images)
        #print("The number of images: ", number_images)

        ## % split for train, val & test
        train_split = round(0.7 * number_images)
        val_split = round(0.15 * number_images)
        test_split = round(0.15 * number_images)   
        #print("Train_split: {}, Val_split: {}, Test_split: {}".format(train_split, val_split, test_split))
        
        
        ## create label folder in train, val & test
        path_label_train = folders[0] + "/" + label
        path_label_val = folders[1] + "/" + label
        path_label_test = folders[2] + "/" + label
        
        os.mkdir(path_label_train)
        os.mkdir(path_label_val)
        os.mkdir(path_label_test)
        
        ## copy each image in train, val & test
        size_train = 0
        for img_train in images[0:train_split]:
            size_train += 1
            shutil.copy2(img_train, path_label_train)
        
        size_val = 0
        for img_val in images[train_split:train_split + val_split]:
            size_val += 1
            shutil.copy2(img_val, path_label_val)
        
        size_test = 0
        for img_test in images[train_split + val_split:]:
            size_test += 1
            shutil.copy2(img_test, path_label_test)
            
        ## add a row of information
        total = size_train + size_val + size_test
        t.add_row([label, size_train, size_val, size_test, total])
        
    print(t)
```

```python
## split the dataset
path = "face_dataset/"
split_dataset(path)
```
